# For the detection of cancerous tumor (malignant)

```python
import pandas as pd
```

```python
dataset = pd.read_csv("cancer.csv")
```

```python
dataset.head(15)
```

## 'x' attributte is going to be our dataset and 'y' attribute is going to be if the tumor is malignant-cancerous(1) or benign-noncancerous(0)

```python
x = dataset.drop(columns = ["diagnosis(1=m, 0=b)"]) # as our x contain only the dataset not the diagnosis about if tumor is(1, 0)
```

```python
y = dataset["diagnosis(1=m, 0=b)"]
```

## now let's split our data into 'training' and 'testing' sets

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2) # here '0.2' refers that 20% of our data is going for testing set  
# and the remaining 80% data we are going to train that
```

### now we will build our AI using tensorflow's keras, so here we will basically be building neural network

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
```

### now we have to start adding layers to our model(like basically we do in neural network)- input, hidden, output layer

## 1st layer - input layer

```python
model.add(tf.keras.layers.Dense(256, input_shape=x_train.shape[1:], activation = 'sigmoid'))
# 1- input layer = 'x_train data', and output = 256 neurons 
# 2- dense layer is a fundamental layer in neural networks that connects every neuron in a layer to every neuron in the previous layer
# 3- sigmoid function always returns output value between 0 and 1 makes it easier to interpret as probability. 
#it helps neural networks make predictions that are easy to understand (as probabilities) and is especially useful for binary classification tasks.
```

Imagine a neural network is like a big team of tiny lights. Each light is a neuron. When you tell this team to do something, like recognize a picture of a cat, the lights work together. Some lights turn on to say, "Hey, I see something fluffy!" while others might say, "I see pointy ears!" Each light talks to its neighbors, and they all help each other to figure out what the picture is. In the end, if enough lights agree that it's a cat, the team tells you, “This is a cat!” So, neurons are like little helpers that work together to solve problems!


## 2nd layer

```python
model.add(tf.keras.layers.Dense(256, activation='sigmoid'))
```

## 3rd layer - output layer


```python
model.add(tf.keras.layers.Dense(1, activation='sigmoid')) 
# here 1 represents that we only need 1 number as an output like if its malignant(1), or benign(0)
```

### now we need to compile the model


```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# optimizer function helps reducing the difference between prediction and actual results(which is called loss)
```

```python
model.fit(x_train, y_train, epochs=948)
# here epochs mean how many times our algorithm is iterating over the same data
# 'fit' is a metos used to train the model
```
